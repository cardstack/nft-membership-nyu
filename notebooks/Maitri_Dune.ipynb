```python
from dunebuggy import Dune
```

```python
dune = Dune()
```

```python
query = dune.fetch_query(2082952)
query.df.head()
```

```python
df2 = query.df.head()
df2
```

```python
import matplotlib.pyplot as plt
```

```python
df2.plot(x="original_currency", y="highest_currency", kind="bar", figsize = (10,5))
plt.xlabel("Original Currency")
plt.ylabel("Percentage")
plt.title("Top 5 Highest Currencies used during Transaction")
# print bar graph
plt.show()
```
